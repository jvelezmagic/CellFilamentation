# 1.0 Cell Filamentation

## Activate project

```julia
using DrWatson
@quickactivate :CellFilamentation
```

## Differential equations algorithms

```julia
choice_function(integrator) = (Int(integrator.dt < 0.001) + 1)
alg_switch = CompositeAlgorithm((Tsit5(), Vern7()), choice_function);
```

## Initialize an Cell structure

```julia
cell = Cell(max_height=4.0)
```

## Simulate filamentation

### Toxin ramp

```julia
t, toxin = toxin_entry(10, 100, 100, 5, false);

Random.seed!(1234)
filamentation_plot = simulation_plot(cell, t, toxin, 100)
```

```julia
savefig(filamentation_plot, plotsdir("filamentation_model_ramp_signal.png"))
```

### Toxin stable

```julia
t2, toxin2 = toxin_entry(10, 100, 100, 1.0, true);

Random.seed!(1234)
filamentation_plot2 = simulation_plot(cell, t2, toxin2, 100)
```

```julia
savefig(filamentation_plot2, plotsdir("filamentation_model_stable_signal.png"))
```
